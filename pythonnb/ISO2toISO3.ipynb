### In this notebook, we convert country codes from iso2 to iso3 to support d3 visualization.

```python
# Load some libraries
import pandas as pd
import json
from pandas import DataFrame
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("../data/data2.csv") # read csv files
df2 = pd.read_csv("../data/data3.csv") 
df = pd.read_csv("../data/all.csv") # read file data containing Iso3 codes
```

### Adding Iso3 codes to the full data.

```python
for i in range (0,len(df1)):
    try:
        df1.ix[i,'id'] = df[df["alpha-2"] == df1.ix[i,'Code']]['alpha-3'].values
    except:
        pass
```

```python
df1.to_csv(path_or_buf="../data/fulldata.csv", index = False)
```
