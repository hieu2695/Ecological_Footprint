```python
# Load some libraries
import pandas as pd
import json
from pandas import DataFrame
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# read excel file
df = pd.read_excel("../data/Data2016.xlsx", header = 2, sheet_name = "Country Graphs 2019 Ed (2016)") 
```

```python
df.head()
```

```python
df = df[['Country','Region','Income Group']] # subset by selecting columns
```

```python
df.head()
```

```python
# change invalid values to NaN
df['Region']=df['Region'].replace('-', np.nan)
df['Income Group']=df['Income Group'].replace([0,'-'], np.nan)
```

```python
df1 = pd.read_csv("../data/data0.csv") # read csv files
df1.reset_index(drop=True,inplace=True)
df2 = pd.read_csv("../data/data1.csv")
df2.reset_index(drop=True,inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
# append region and income group to df1
# try, except to ignore missing data from World and continents
for i in range (0,len(df1)):
    try:
        df1.ix[i,'Region'] = df[df.Country == df1.ix[i,'Country']]['Region'].values
        df1.ix[i,'Income'] = df[df.Country == df1.ix[i,'Country']]['Income Group'].values
    except:
        pass
```

```python
# do the same for df2
for i in range (0,len(df2)):
    try:
        df2.ix[i,'Region'] = df[df.Country == df2.ix[i,'Country']]['Region'].values
        df2.ix[i,'Income'] = df[df.Country == df2.ix[i,'Country']]['Income Group'].values
    except:
        pass
```
